```python

#Execicio 7 "	Calculadora simples Peça dois números e uma operação (+, -, *, /). Use if/elif para calcular o resultado. "
print("--- Calculadora Simples ---")
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))

operacao = input("Digite a operação (+, -, *, /): ")
if operacao == '+':
    resultado = num1 + num2
    print(f"O resultado de {num1} + {num2} é: {resultado}")
elif operacao == '-':
    resultado = num1 - num2
    print(f"O resultado de {num1} - {num2} é: {resultado}")
elif operacao == '*':
    resultado = num1 * num2
    print(f"O resultado de {num1} * {num2} é: {resultado}")
elif operacao == '/':
    if num2 != 0:
        resultado = num1 / num2
        print(f"O resultado de {num1} / {num2} é: {resultado}")
    else:
        print("Erro: Divisão por zero não é permitida.")
else:
    print("Operação inválida. Por favor, use +, -, *, ou /.")
```

```python

#Execicio 8 "	Dia da semana Peça um número de 1 a 7 e mostre o dia correspondente (ex: 1 = domingo). "

print("--- Qual é o dia da semana? ---")
numero_dia = int(input("Digite um número de 1 a 7: "))
if numero_dia == 1:
    print("1 = Domingo")
elif numero_dia == 2:
    print("2 = Segunda-feira")
elif numero_dia == 3:
    print("3 = Terça-feira")
elif numero_dia == 4:
    print("4 = Quarta-feira")
elif numero_dia == 5:
    print("5 = Quinta-feira")
elif numero_dia == 6:
    print("6 = Sexta-feira")
elif numero_dia == 7:
    print("7 = Sábado")
else:
    print("Número inválido. Por favor, digite um número entre 1 e 7.")
```

```python

#Execicio 9 "	Ano bissexto Solicite um ano e informe se ele é bissexto ou não. "
print("--- Verificador de Ano Bissexto ---")

# Solicita o ano ao usuário
ano = int(input("Digite um ano: "))

# Verifica se o ano é bissexto
# Um ano é bissexto se for divisível por 4, exceto se for divisível por 100
# Mas é bissexto se for divisível por 400
if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print(f"O ano {ano} é um ano bissexto.")
else:
    print(f"O ano {ano} NÃO é um ano bissexto.")
```
